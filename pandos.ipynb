```python
m14 = d14small[d14small['lytis'] == 'M']
m14.head(7)
```

```python
# koks buvo vyrÅ³ ir moterÅ³ vidutinis atlyginimas 2014 metais?
# rezulataÄ… iÅ¡spausdinkite vienas Å¡alia kito:
# pvz 
# VyrÅ³ x, moterÅ³ y atlyginimai
m14vid = m14['bdu_spalio'].mean()
f14vid = f14['bdu_spalio'].mean()
print('Vyru vidutinis atlyginimas:', m14vid, 'Moteru vidutinis atlyginimas:', f14vid, sep= '\n')
```

```python
# and ir or operatoriai:
# and = &
# or = |
# atrinkime tik vyrus, kuriu issilavinimas pazymetas G3
m14g3 = d14small[(d14small['lytis'] == 'M') & (d14small['issilavinimas'] == 'G3')]
m14g3.head()
```

```python
# atrinkite tik vyrus
#pagal issilavinima
# G1, G2, G3, G4
#ir suraskite ju vidutinius atlyginimus
# spausdinkite taip:
# G1 x
# G2 y
# G3 z
# G4 e
m14g1 = d14small[(d14small['lytis'] == 'M') & (d14small['issilavinimas'] == 'G1')]
m14g2 = d14small[(d14small['lytis'] == 'M') & (d14small['issilavinimas'] == 'G2')]
m14g3 = d14small[(d14small['lytis'] == 'M') & (d14small['issilavinimas'] == 'G3')]
m14g4 = d14small[(d14small['lytis'] == 'M') & (d14small['issilavinimas'] == 'G4')]
m14g1vid = m14g1['bdu_spalio'].mean()
m14g2vid = m14g2['bdu_spalio'].mean()
m14g3vid = m14g3['bdu_spalio'].mean()
m14g4vid = m14g4['bdu_spalio'].mean()
print('G1 x:', m14g1vid)
print('G2 y:', m14g2vid)
print('G3 z:', m14g3vid)
print('G4 e:', m14g4vid)
```
